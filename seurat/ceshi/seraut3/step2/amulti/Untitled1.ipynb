```r
suppressMessages({
library(ggplot2)
library(reshape2)
library(argparser)
library(Seurat)
library(corrplot)
library(dplyr)
library(grid)
library(cowplot)
library(dplyr)
library(grid)
library(tidyverse)
})
```

```r
RAW1 <- read.table("/SGRNJ02/RandD4/RD2019030_SCOPEv2/20201008/S091802-4/05.count/S091802-4_matrix.tsv.gz",sep = "\t",header=TRUE,row.names=1)
```

```r
PRO1 <- CreateSeuratObject(counts = RAW1, project = 'S091802-4', min.cells = 5)
```

```r
PRO1 <- RenameCells(PRO1, add.cell.id = "S091802-4")
```

```r
PRO1[["percent.mt"]] <- PercentageFeatureSet(PRO1, pattern = "^MT-")
```

```r
head(PRO1@meta.data,5)
```

```r
VlnPlot(object = PRO1, features = c("nFeature_RNA", "nCount_RNA", "percent.mt"), ncol = 3)
```

```r
PRO1 <-  NormalizeData(object = PRO1)
PRO1 <-  ScaleData(object = PRO1)
```

```r
PRO1 <- FindVariableFeatures(object = PRO1)
genes.use<- head(HVFInfo(object = PRO1),2000)
```

```r
PRO <- RunPCA(object=PRO1,features = VariableFeatures(object = PRO1))
PRO <- FindNeighbors(PRO, reduction = "pca", dims = 1:20)
```

```r
PRO <- FindClusters(PRO,resolution = 0.8, algorithm = 1)
PRO <- RunTSNE(object=PRO,dims.use=1:20,do.fast=TRUE,check_duplicates = FALSE)
PRO <- RunUMAP(PRO, reduction = "pca", dims = 1:20)
```

```r
DimPlot(object = PRO, reduction = "umap",label = TRUE)
```

```r
freq_table <- prop.table(x=table(PRO@active.ident ,PRO@meta.data[,"orig.ident"]),margin=2)
write.table(freq_table,file='123')
```

```r
dat <- melt(freq_table,varnames=c("type","orig.ident"),value.name = "percent")
dat$type <- as.character(dat$type)
```

```r
ggplot(dat, aes(x = orig.ident, y = percent,fill = type )) + geom_bar(stat = 'identity') + coord_flip()
```
